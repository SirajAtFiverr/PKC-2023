```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading dataset
iris  = sns.load_dataset('iris')
iris.head()
```

## Data Visualization using Matplotlib

```python
### Plot style
plt.style.use('ggplot')
```

### Scatter Plot

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10), sharex=True)

ax[0][0].scatter(iris['sepal_length'], iris['sepal_width'])
ax[0][0].set_title('Sepal Length vs Sepal Width')

ax[0][1].scatter(iris['petal_length'], iris['petal_width'])
ax[0][1].set_title('Petal Length vs Petal Width')

ax[1][0].scatter(iris['sepal_length'], iris['petal_length'])
ax[1][0].set_title('Sepal Length vs Petal Length')

ax[1][1].scatter(iris['sepal_width'], iris['petal_width'])
ax[1][1].set_title('Sepal Width vs Petal Width');
```

```python
colors = {'setosa':'r', 'versicolor':'g', 'virginica':'b'}
c = list(map(lambda x: colors[x], iris['species']))
plt.scatter(iris['sepal_length'], iris['sepal_width'], c=c, marker='x', s=50);
```

### Bar Plot

```python
plt.bar("Sepal Length", iris['sepal_length'].mean(), yerr=iris['sepal_length'].std());
```

```python
plt.bar(iris['species'], iris['petal_length']);
```

```python
plt.errorbar(iris['species'], iris['petal_length'], yerr=iris['petal_length'].std());
```

### Histogram

```python
plt.hist(iris['sepal_length'], bins=20, color='green', edgecolor='white', linewidth=2, label="sepal_length");
plt.hist(iris['sepal_width'], bins=20, color='blue', edgecolor='white', linewidth=2, label="sepal_width");
plt.hist(iris['petal_length'], bins=20, color='red', edgecolor='white', linewidth=2, label="petal_length");
plt.hist(iris['petal_width'], bins=20, color='yellow', edgecolor='white', linewidth=2, label="petal_width");
plt.legend()
plt.show();
```

### Box Plot

```python
plt.boxplot(iris["petal_width"], showmeans=True, meanline=True);
plt.show();
```

### Pie Chart

```python
plt.pie(iris['species'].value_counts(), labels=iris['species'].unique(),
        autopct='%1.1f%%', radius=1.2, explode=[0.01, 0.01, 0.01]);
```

### Violin Plot

```python
plt.violinplot(iris['petal_length'], showmeans=True, showmedians=True);
```

## Data Visualization using Seaborn 

### Builtin Datasets

```python
sns.get_dataset_names()
```

```python
# LOAD DATASET
tips = sns.load_dataset('tips')
tips.head()
```

```python
# changing pallete
sns.set_palette('husl')
# changing style
sns.set_style('darkgrid')
```

### Scatter Plot

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='time', style='day', size='size');
```

### Line Plot

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=iris, hue='species', errorbar=None);
```

### Relational Plot (scatter & line)

```python
sns.relplot(x='sepal_length', y='sepal_width', data=iris, hue='species', kind='line', errorbar=None);
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', kind='scatter', col='time', row='day');
```

### Count Plot

```python
sns.countplot(x='sex', data=tips);
```

### Bar Plot

```python
sns.barplot(x="time", y="tip", hue="sex", data=tips);
```

### Point Plot

```python
sns.pointplot(y='total_bill', x="time", data=tips, capsize=.2);
```

### Box Plot

```python
sns.boxplot(x="day", y="tip", hue="sex", data=tips);
```

### Boxen Plot

```python
sns.boxenplot(x="day", y="tip", hue="sex", data=tips);
```

```python
sns.violinplot(x="day", y="tip", hue="sex", data=tips);
```

### Strip Plot

```python
sns.stripplot(x="day", y="tip", hue="sex", data=tips);
```

### Swarm Plot

```python
sns.swarmplot(x="day", y="tip", hue="sex", data=tips);
```

### Catplot (countplot, barplot, boxplot, violinplot, stripplot, swarmplot)

```python
sns.catplot(x="day", y="tip", hue="sex", data=tips, kind="bar", col="time", row="smoker");
```

### Pallete Plot

```python
sns.palplot(sns.color_palette("husl", 10));
```

### Histogram

```python
sns.histplot(x="total_bill", data=tips, kde=True, bins=10);
```

### Karnal Density Plot

```python
sns.kdeplot(x="total_bill", data=tips, fill=True);
```

### ECDF Plot

```python
sns.ecdfplot(x="total_bill", data=tips);
```

### Rug Plot

```python
sns.displot(x="total_bill", data=tips, kind="kde", fill=True, rug=True);
```

### Dis Plot (hist, kde, rug, ecdf)

```python
sns.displot(x="total_bill", data=tips, kde=True, bins=10, col="time");
```

### Regression Plot

```python
sns.regplot(x="total_bill", y="tip", data=tips, col="time");
```

### Residual Plot

```python
sns.residplot(x="total_bill", y="tip", data=tips);
```

### Linear Model Plot (regplot, residplot)

```python
sns.lmplot(x="total_bill", y="tip", data=tips, col="time", row='sex');
```

### Heatmap

```python
new_iris = iris.drop('species', axis=1)
```

```python
import numpy as np
sns.heatmap(new_iris.corr(), annot=True, cmap='coolwarm', mask=np.triu(new_iris.corr(), k=1));
```

### Facet Grid  
catplot, relplot, lmplot, and displot are interface for using FacetGrid

```python
g = sns.FacetGrid(tips, col="time", row="smoker");
g.map(sns.scatterplot, 'total_bill', 'tip');
```

### Pair Grid

```python
g = sns.PairGrid(iris, vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], hue='species');
g.map_diag(sns.histplot);
g.map_offdiag(sns.scatterplot);
```

### Pair Plot

```python
sns.pairplot(iris, hue='species', diag_kind='hist', corner=True);
```

### Joint Grid

```python
g = sns.JointGrid(data=tips, x="total_bill", y="tip");
g.plot_joint(sns.scatterplot);
g.plot_marginals(sns.histplot, kde=True);
```

### Joint Plot

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter", marginal_kws=dict(bins=20, kde=True));
```
